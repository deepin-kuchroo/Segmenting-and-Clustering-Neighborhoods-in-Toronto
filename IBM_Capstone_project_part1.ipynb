# Coursera Applied Data Science Capstone Project Week 3 Segmenting and Clustering Neighborhoods in Toronto Part 1

 installing required libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

scrape data set from Wikipedia website

```python

# read the webpage from the wiki
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
#Process and convert html data
data = response.text
soup = BeautifulSoup(data,'html.parser')
wiki_table=soup.find('table')
#develop dataframe
df = pd.read_html(str(wiki_table))[0]
df.head()
#df.shape
```

prepare the dataframe in pursuant of instructions

```python
#Rename the columns names
df.rename(columns={0:'Postalcode', 1: 'Borough', 2: 'Neighborhood'}, inplace=True)
df.head()
```

```python
# Remove "Borough" with 'Not assigned' values
df2=df[df['Borough'].str.contains("Not assigned") == False].reset_index()
df2.head()
```

```python
df3= df2.groupby(['Postalcode', 'Borough'])['Neighborhood'].apply(', '.join).reset_index()
df3.head()
```

Get the number of rows of dataframe

Save the file as csv file for using in part 2

```python
#Save data as'Capstone_part1.csv'
df3.to_csv('Capstone_part1.csv')
print('Successfully Saved!')
```
